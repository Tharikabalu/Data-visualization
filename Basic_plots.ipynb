```python
import matplotlib.pyplot as plt
import numpy as np
```

Line Plot

```python
x=np.array([1,2,3,4])
y=x*2
plt.plot(x,y)
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
plt.plot(x1, y1, '-.')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

Bar Plot

```python
data = {'C':20, 'C++':15, 'Java':30, 'Python':35}
courses = list(data.keys())
values = list(data.values())
plt.bar(courses,values)
plt.show()
```

Horizontal Bar Plot


```python
data = {'C':20, 'C++':15, 'Java':30, 'Python':35}
courses = list(data.keys())
values = list(data.values())
fig,ax=plt.subplots(figsize=(8,4))
ax.barh(courses,values)
plt.show()
```

Multiple bar plots

```python
year=[2020,2021,2022,2023,2024]
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]
fig,ax=plt.subplots(1,3,figsize=(15,5))
ax[0].bar(year,IT)
ax[1].bar(year,ECE)
ax[2].bar(year,CSE)
ax[0].set_title("IT")
ax[1].set_title("ECE")
ax[2].set_title("CSE")

plt.show()
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(6,4))

IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]
year=[2020,2021,2022,2023,2024]

br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1,IT,width=0.25,color='r')
plt.bar(br2,ECE,width=0.25,color='g')
plt.bar(br3,CSE,width=0.25,color='b')

plt.xticks(br2,year)

plt.show()


```

Stacked Bar

```python
boys=[40,35,70,15,20]
ind1=np.arange(len(boys))
plt.bar(ind1,boys,width=0.5,label='Boys')
plt.bar(ind1,100-np.array(boys),width=0.5,bottom=boys,label='Girls')
plt.xticks(ind1,labels=['T1','T2','T3','T4','T5'])
plt.legend()
total=np.array(boys)+(100-np.array(boys))
for i, t in enumerate(total):
    plt.text(ind1[i]-0.15, boys[i]-5, f'{boys[i]}%',color='white', fontsize=10)
    plt.text(ind1[i]-0.15, t - 5, f'{100 - boys[i]}%',color='white', fontsize=10)
plt.show()
```

Pie Chart

```python
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
data = [30, 20, 10, 15, 5, 5]

plt.figure(figsize=(8, 8))

plt.pie(data, labels=cars, autopct='%1.1f%%')
plt.legend(cars, loc='upper right')
plt.axis('equal')

plt.show()
```

Scatter Plot

```python
x1 = [89, 43, 36, 36, 95, 10,66, 34, 38, 20]
y1 = [21, 46, 3, 35, 67, 95,33, 72, 58, 10]

x2 = [26, 29, 48, 64, 6, 5,36, 66, 72, 40]
y2 = [26, 34, 90, 33, 38,20, 56, 2, 47, 15]

plt.scatter(x1,y1,marker="s",color='blue',label='Data1')
plt.scatter(x2,y2,marker="^",color='red',label='Data2')
plt.legend()
plt.show()
```

```python
ax=plt.axes(projection='3d')
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
ax.plot3D(x,y,z)
plt.show()
```

```python
ax = plt.axes(projection ='3d')
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = np.random.rand(num_points)
ax.scatter(x, y, z)
plt.show()
```

```python
ax = plt.axes(projection ='3d')
x = np.outer(np.linspace(-2, 2, 10), np.ones(10))
y = x.copy().T
z = np.cos(x ** 2 + y ** 3)
ax.plot_surface(x,y,z,cmap='viridis')
plt.show()
```
